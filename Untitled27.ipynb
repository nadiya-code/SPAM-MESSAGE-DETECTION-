```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
import re
import string
```

```python
df = pd.read_csv("spam.csv", dtype=str)

# Standardize column names
df.columns = [c.lower().strip() for c in df.columns]

```

```python
df.columns = [c.lower().strip() for c in df.columns]

print("\nCOLUMNS FOUND:", df.columns)
```

```python
# Drop rows missing message OR label
df = df.dropna(subset=["message", "label"])
```

```python
# Replace blanks and weird whitespace with NaN
df = df.replace(["", " ", "  ", "   ", "    ", None], np.nan)

```

```python

# Convert to string
df["message"] = df["message"].astype(str)
df["label"] = df["label"].astype(str)
```

```python
# Remove empty messages
df = df[df["message"].str.strip() != ""]

print("\nAfter cleaning:", df.shape)
```

```python
df.head(10)
```

```python
try:
    df["label_num"] = df["label"].astype(int)
except:
    # In case dataset contains text labels
    df["label_num"] = df["label"].map({"ham": 0, "spam": 1})

print("\nLabel distribution:")
print(df["label_num"].value_counts())
```

```python
def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = text.translate(str.maketrans("", "", string.punctuation))
    text = re.sub(r"\d+", "", text)
    return text.strip()

df["clean_text"] = df["message"].astype(str).apply(clean_text)

print("\nSample Cleaned Messages:")
print(df["clean_text"].head())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x=df['label_num'])
plt.title("Spam vs Ham Count")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df["clean_text"],
    df["label_num"],
    test_size=0.2,
    random_state=42,
    stratify=df["label_num"]    # avoids errors with imbalanced data
)
```

```python
tfidf = TfidfVectorizer(stop_words="english", max_features=5000)

X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf  = tfidf.transform(X_test)
```

```python
models = {
    "Naive Bayes": MultinomialNB(),
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM": LinearSVC(),
    "Random Forest": RandomForestClassifier()
}
```

```python
results = {}
```

```python
for name, model in models.items():
    print(f"\n====== {name} ======")
    model.fit(X_train_tfidf, y_train)
    preds = model.predict(X_test_tfidf)
    acc = accuracy_score(y_test, preds)
    results[name] = acc
    print("Accuracy:", acc)
    print(classification_report(y_test, preds))
```

```python
best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
print(" BEST MODEL:", best_model_name)
print(" Accuracy:", results[best_model_name])
```

```python
def predict_message(msg):
    msg_clean = clean_text(msg)
    vec = tfidf.transform([msg_clean])
    pred = best_model.predict(vec)[0]
    return "SPAM" if pred == 1 else "HAM"
```

```python
print("\nPrediction Examples:")
print("Example 1:", predict_message("You won a FREE iPhone! Click here"))
print("Example 2:", predict_message("Hey, let's meet tomorrow"))
```

```python
from sklearn.metrics import confusion_matrix


# CONFUSION MATRIX FOR BEST MODEL
y_pred_best = best_model.predict(X_test_tfidf)

cm = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Ham", "Spam"], yticklabels=["Ham", "Spam"])
plt.title(f"Confusion Matrix - {best_model_name}")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# ACCURACY COMPARISON GRAPH
plt.figure(figsize=(8,5))
plt.bar(results.keys(), results.values(), color='skyblue')
plt.title("Accuracy Comparison of ML Models")
plt.ylabel("Accuracy")
plt.ylim(0, 1)  # Scale between 0 and 1
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve

# Get decision function or probabilities
try:
    y_scores = best_model.decision_function(X_test_tfidf)
except:
    y_scores = best_model.predict_proba(X_test_tfidf)[:, 1]

precision, recall, thresholds = precision_recall_curve(y_test, y_scores)

plt.figure(figsize=(7,5))
plt.plot(recall, precision, label="Precision-Recall Curve")
plt.title(f"Precision-Recall Curve ({best_model_name})")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

try:
    y_scores = best_model.decision_function(X_test_tfidf)
except:
    y_scores = best_model.predict_proba(X_test_tfidf)[:, 1]

fpr, tpr, thresholds = roc_curve(y_test, y_scores)
auc = roc_auc_score(y_test, y_scores)

plt.figure(figsize=(7,5))
plt.plot(fpr, tpr, label=f"AUC = {auc:.4f}")
plt.plot([0,1], [0,1], linestyle="--", color="gray")
plt.title(f"ROC Curve ({best_model_name})")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import time

results = {}
times = {}

for name, model in models.items():
    print(f"\n====== {name} ======")
    
    start = time.time()
    model.fit(X_train_tfidf, y_train)
    end = time.time()

    preds = model.predict(X_test_tfidf)
    acc = accuracy_score(y_test, preds)
    
    results[name] = acc
    times[name] = end - start

    print("Accuracy:", acc)
    print("Training Time:", round(times[name], 4), "seconds")
    print(classification_report(y_test, preds))
```

```python
plt.figure(figsize=(8,5))
plt.bar(times.keys(), times.values(), color='orange')
plt.title("Training Time Comparison")
plt.ylabel("Time (seconds)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
